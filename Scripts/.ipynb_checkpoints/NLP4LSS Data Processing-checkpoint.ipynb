```python
import os
import pandas as pd

# Base path
base_path = r'C:\Users\sakar\Downloads\NLP4LSS Project'

# Specify the target file or directory
target_path = os.path.join(base_path, 'NLP4LSS Data\hein-daily')

# Matching string patterns
speeches_files_pattern = 'speeches'
speakermap_files_pattern = 'SpeakerMap'

# Get the names of files matching the string pattern
speeches = [
    file for file in os.listdir(target_path) if speeches_files_pattern in file
]

speakermaps = [
    file for file in os.listdir(target_path) if speakermap_files_pattern in file
]
```

```python
# Read and concatenate the speeches files
speeches_data_frames = []
for file_name in speeches:
    file_path = os.path.join(target_path, file_name)
    try:
        df = pd.read_csv(file_path, delimiter='|', error_bad_lines=False, encoding='latin-1')
        speeches_data_frames.append(df)
    except pd.errors.ParserError:
        print(f"Skipped file '{file_name}' due to parsing error.")

speeches_concatenated = pd.concat(speeches_data_frames)
```

```python
# Read and concatenate the speakermap files
speakermaps_data_frames = []
for file_name in speakermaps:
    file_path = os.path.join(target_path, file_name)
    try:
        if file_name.endswith('.txt'):
            df = pd.read_csv(file_path, delimiter='|', encoding='utf-8')
        elif file_name.endswith('.csv'):
            df = pd.read_csv(file_path, encoding='utf-8')
        speakermaps_data_frames.append(df)
    except pd.errors.ParserError:
        print(f"Skipped file '{file_name}' due to parsing error.")

speakermaps_concatenated = pd.concat(speakermaps_data_frames)
```

```python
sorted_speakermaps_concatenated = speakermaps_concatenated.sort_values(by='speakerid', ascending=False)
```

```python
# Merge the DataFrames based on the 'speech_id' column
merged_df = sorted_speakermaps_concatenated.merge(speeches_concatenated, on='speech_id')
```

```python
# Specify the output file pat
output_path = os.path.join(base_path, 'NLP4LSS Data\merged_data.csv')

# Save the merged DataFrame to a CSV file
merged_df.to_csv(output_path, index=False)
```

```python
# Initialize empty dictionaries
speaker_name_dict = {}
speeches_dict = {}

index = 0

# Iterate over the DataFrame rows
for row in merged_df.itertuples(index=False):
    speech_id = row.speakerid
    speaker_name = f"{row.firstname} {row.lastname}"
    speech = str(row.speech)
    
    index = index + 1
    print(str(index) + ' of ' + str(2583087))
    
    # Check if the speaker ID exists in the dictionaries
    if speech_id in speaker_name_dict:
        # Append the speech to the existing value in speeches_dict
        speeches_dict[speech_id] += '<new>' + speech
    else:
        # Create a new entry in both dictionaries
        speaker_name_dict[speech_id] = speaker_name
        speeches_dict[speech_id] = speech
```

```python
# Specify the output file paths
output_txt_path_speeches = os.path.join(base_path, 'NLP4LSS Data', 'speeches_dict.txt')
output_txt_path_speakers = os.path.join(base_path, 'NLP4LSS Data', 'speakers_dict.txt')

index = 0

# Open the speeches file in write mode
with open(output_txt_path_speeches, 'w', encoding='utf-8') as file_speeches:
    index = index + 1
    print(str(index))
    # Write speeches_dict
    file_speeches.write("Speeches:\n")
    for speech_id, speeches in speeches_dict.items():
        file_speeches.write(f"Speaker ID: {speech_id}\n")
        file_speeches.write(f"Speeches: {speeches}\n")
        file_speeches.write('\n')

index = 0
        
# Open the speakers file in write mode
with open(output_txt_path_speakers, 'w', encoding='utf-8') as file_speakers:
    index = index + 1
    print(str(index))
    # Write speaker_name_dict
    file_speakers.write("Speaker Names:\n")
    for speaker_id, speaker_name in speaker_name_dict.items():
        file_speakers.write(f"Speaker ID: {speaker_id}\n")
        file_speakers.write(f"Speaker Name: {speaker_name}\n")
        file_speakers.write('\n')
```

```python
import random

# Specify the number of random speakers to select
num_random_speakers = 20

# Get all speaker IDs from the speaker_name_dict
speaker_ids = list(speaker_name_dict.keys())

# Randomly select num_random_speakers IDs from the speaker IDs
random_speaker_ids = random.sample(speaker_ids, num_random_speakers)

# Create a sub-dictionary of random speakers and speeches with the same IDs
random_dict = {}
for speaker_id in random_speaker_ids:
    speaker_name = speaker_name_dict.get(speaker_id, "")
    speeches = speeches_dict.get(speaker_id, "")
    random_dict[speaker_id] = (speaker_name, speeches)

# Print the random speakers and their speeches
#print("Random Speakers and Speeches:")
#for speaker_id, (speaker_name, speeches) in random_dict.items():
#    print(f"Speaker ID: {speaker_id}, Speaker Name: {speaker_name}, Speeches: {speeches}")
```

```python
# Specify the output file path
output_txt_path = os.path.join(base_path, 'NLP4LSS Data', 'random_dict.txt')

# Open the file in write mode
with open(output_txt_path, 'w', encoding='utf-8') as file:
    # Write the random dictionary to the file
    file.write("Random Speakers and Speeches:\n")
    for speaker_id, (speaker_name, speeches) in random_dict.items():
        file.write(f"Speaker ID: {speaker_id}\n")
        file.write(f"Speaker Name: {speaker_name}\n")
        file.write(f"Speeches: {speeches}\n")
        file.write('\n')
```

```python
# Specify the paths of the speaker and speeches files
input_txt_path_speakers = os.path.join(base_path, 'NLP4LSS Data', 'speakers_dict.txt')
input_txt_path_speeches = os.path.join(base_path, 'NLP4LSS Data', 'speeches_dict.txt')

# Initialize empty dictionaries
speaker_name_dict = {}
speeches_dict = {}

# Read the speakers file and populate the speaker_name_dict dictionary
with open(input_txt_path_speakers, 'r', encoding='utf-8') as file_speakers:
    lines = file_speakers.readlines()
    speaker_id = None
    speaker_name = None
    for line in lines:
        if line.startswith("Speaker ID:"):
            speaker_id = line.strip().split(":")[1].strip()
        elif line.startswith("Speaker Name:"):
            speaker_name = line.strip().split(":")[1].strip()
            if speaker_id and speaker_name:
                speaker_name_dict[speaker_id] = speaker_name
                speaker_id = None
                speaker_name = None

# Read the speeches file and populate the speeches_dict dictionary
with open(input_txt_path_speeches, 'r', encoding='utf-8') as file_speeches:
    lines = file_speeches.readlines()
    speaker_id = None
    speeches = None
    for line in lines:
        if line.startswith("Speaker ID:"):
            speaker_id = line.strip().split(":")[1].strip()
        elif line.startswith("Speeches:"):
            speeches = line.strip().split(":")[1].strip()
            if speaker_id and speeches:
                speeches_dict[speaker_id] = speeches
                speaker_id = None
                speeches = None

# Print the dictionaries
print("Speaker Names:")
print(speaker_name_dict)
print("\nSpeeches:")
print(speeches_dict)
```
